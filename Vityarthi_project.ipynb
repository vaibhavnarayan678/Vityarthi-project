```python
import time

def phi(n):
    result = n
    if n % 2 == 0:
        result -= result // 2
        while n % 2 == 0:
            n //= 2
    p = 3
    while p * p <= n:
        if n % p == 0:
            result -= result // p
            while n % p == 0:
                n //= p
        p += 2
    if n > 1:
        result -= result // n
    return result

n = int(input("Enter first number: "))
p = int(input("Enter second number: "))

start = time.time()
print("phi(", n, ") =", phi(n))
print("Time:", time.time() - start, "seconds")

start = time.time()
print("phi(", p, ") =", phi(p))
print("Time:", time.time() - start, "seconds")

print("Done!")
```

```python
import time

def mu(n):
    if n == 1:
        return 1
    count = 0
    i = 2
    while i * i <= n:
        if n % i == 0:
            n = n // i
            if n % i == 0:
                return 0
            count = count + 1
        i = i + 1
    if n > 1:
        count = count + 1
    if count % 2 == 0:
        return 1
    else:
        return -1

print("MOBIUS FUNCTION CALCULATOR")

n = int(input("enter n : "))
p = int(input("enter p : "))

print("\ncalculating...")

start = time.time()
print("mu(", n, ") =", mu(n))
print("time:", time.time() - start, "sec")

start = time.time()
print("mu(", p, ") =", mu(p))
print("time:", time.time() - start, "sec")

print("\nfinished thank you sir")
```

```python
import time
import math

def prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def pi(n):
    count = 0
    for i in range(2, n + 1):
        if prime(i):
            count = count + 1
    return count

print("PRIME COUNTING FUNCTION")


n = int(input("enter n : "))
p = int(input("enter p : "))

print("\ncounting primes...")

start = time.time()
print("pi(", n, ") =", pi(n))
print("time taken:", time.time() - start, "seconds")

start = time.time()
print("pi(", p, ") =", pi(p))
print("time taken:", time.time() - start, "seconds")

print("\ndone sir please check")
```

```python
import time

def power(a, b, m):
    res = 1
    a = a % m
    while b > 0:
        if b % 2 == 1:
            res = (res * a) % m
        a = (a * a) % m
        b = b // 2
    return res

def legendre(n, p):
    n = n % p
    if n == 0:
        return 0
    x = power(n, (p-1)//2, p)
    if x == 1:
        return 1
    if x == p-1:
        return -1
    return 0

print("LEGENDARE SYMBOL CALCULATOR")


n = int(input("enter n : "))
p = int(input("enter prime p : "))

start = time.time()
ans = legendre(n, p)
print("Legendre(", n, "/", p, ") =", ans)
print("time:", time.time() - start, "sec")

print("thank you sir")
```

```python
n = int(input("how many equations: "))

a = []
m = []

for i in range(n):
    print("equation", i+1)
    x = int(input("remainder: "))
    y = int(input("modulus: "))
    a.append(x)
    m.append(y)

M = 1
for i in m:
    M = M * i

total = 0
for i in range(n):
    Mi = M // m[i]
    inv = 1
    while (Mi * inv) % m[i] != 1:
        inv = inv + 1
    total = total + a[i] * Mi * inv

ans = total % M
print("answer =", ans)
print("done")
```
